### EDA and Feature Engineering - Flight Price Prediction

```python
# Import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.preprocessing import OneHotEncoder
```

```python
# Read execl file into a dataframe
df=pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.tail()
```

```python
# Get basic info about data
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# Feature engineering for Date_of_Journey
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
# Convert the date columns as int type from object
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
# Drop Date_of_Journey column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
# Feature engineering for Arrival_Time
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
# Split hour and minute for Arrival_Time
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
# Convert Arrival_hour and Arrival_Minute columns as int type from object
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_Minute']=df['Arrival_Minute'].astype(int)
```

```python
df.info()
```

```python
# Drop Arrival_Time column
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
# Feature engineering on Dep_Time
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_minute']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
# Convert Departure_hour and Departure_minute columns as int type from object
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_minute']=df['Departure_minute'].astype(int)
```

```python
# Drop Dep_Time column
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head(2)
```

```python
# Feature engineering on Total_Stops
df['Total_Stops'].unique()
```

```python
# Check if Total_Stops have any null values
df['Total_Stops'].isnull().sum()
```

```python
# Convert stops into numerical data and replace nan with 1.
df['Total_Stops']=df['Total_Stops'].map({
    'non-stop':0,
    '1 stop':1,
    '2 stops':2,
    '3 stops':3,
    '4 stops':4,
    np.nan:1
    })
```

```python
# Check if Total_Stops have any null values
df['Total_Stops'].isnull().sum()
```

```python
# Drop Route column.
df.drop('Route',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
# Feature engineering on Duration
extracted=df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int) + df['Duration'].str.extract(r'(\d+)m').fillna(0).astype(int) / 60
```

```python
# Change datatype to float of the 'extracted'.
df['Duration']=extracted.astype(float)
```

```python
# Check if Duration have any null values
df['Duration'].isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
# Check unique values in Airline.
df['Airline'].unique()
```

```python
# Check unique values in Source.
df['Source'].unique()
```

```python
# Check unique values in Additional_Info.
df['Additional_Info'].unique()
```

```python
# Check unique values in Destination.
df['Destination'].unique()
```

```python
# Creating instance for OneHotEncoding.
encoder=OneHotEncoder()
```

```python
# Converting into numerical data
encoder.fit_transform(df[['Airline','Source','Additional_Info','Destination']]).toarray()
```

```python
# Converting numerical data into DataFrame
Encoded_data=pd.DataFrame(encoder.fit_transform(df[['Airline','Source','Additional_Info','Destination']]).toarray(),columns=encoder.get_feature_names_out())
```

```python
Encoded_data
```

```python
# Joining the encoded DataFrame with the main DataFrame
df=pd.concat([df,Encoded_data],axis=1)
df
```

```python
df.info()
```

```python
# Drop Airline column.
df.drop('Airline',inplace=True,axis=1)
```

```python
# Drop Source column.
df.drop('Source',inplace=True,axis=1)
```

```python
# Drop Additional_Info column.
df.drop('Additional_Info',inplace=True,axis=1)
```

```python
# Drop Destination column.
df.drop('Destination',inplace=True,axis=1)
```
